```python
from datetime import datetime
# Date Time module is used to get the current date and time
#class method is used for updating and displaying the template of the student marksheet
now = datetime.now()
x = now.strftime("%d/%m/%y, %H:%M:%S")
class Student:
  def __init__(self,details): # self - instance variable
    # class constructor
    #Contains the reference of details provided
    self.name=details[0]
    self.rollno=details[1]
    self.programme=details[2]
    self.year=details[3]
    self.marks1=details[4]
    self.marks2=details[5]
    self.marks3=details[6]
    self.marks4=details[7]
    self.marks5=details[8]
    
  
  
  def updateName(self,newName):
    # this method is used for updating the name 
        self.name=newName
        return str(self)
     
  def updateProgramme(self,newProgramme):
    # this method is used for updating the programme name  
        self.programme=newProgramme
        return str(self)
    
  def updateMarks(self,newMarks,sub):
    # this method is used for updating the marks
        if sub==1:
          self.marks1=newMarks
        elif sub==2:
          self.marks2=newMarks
        elif sub==3:
          self.marks3=newMarks
        elif sub==4:
          self.marks4=newMarks
        elif sub==5:
          self.marks5=newMarks
        
        return str(self)

  def  updateYear(self,newYear):
   #this method used for updating the passing year 
    self.year=newYear
    return str(self)
  #def updatedoi(self):
    #self.doi = newdoi
    #newdoi = datetime.now()
    #return newdoi
    
  def template(self):
    # this method stores the template of the marksheet
    template=f'''     University of Delhi
            Delhi
Student name: {self.name}
Programme: {self.programme}
Roll Number: {self.rollno}

Sub1: {self.marks1}
Sub2: {self.marks2}
Sub3: {self.marks3}
Sub4: {self.marks4}
Sub5: {self.marks5}

Year of passing: {self.year}


Date of Issue: {x}

Signing Authority '''

    return template
```

```python
def myMSGenerator(fileName):
  # This method is used for generating the marksheet using the filename
  try:                                    
      with open(fileName, "r") as f:      #opening a file in read mode
        for line in f.readlines()[2:]:
            data = line.strip().split(", ")
            student = Student(data)         #Creating object 
            with open(f"{student.name}_{student.rollno}.txt", "w") as ms:
              ms.write(student.template())    #Generate Marksheet

      print("Marksheets are generated successfully!!")
  except FileNotFoundError :
          print(f"Error: File Not Found invalid file")
  except Exception as e:
          print("Error: ",e)
```

```python
def myMSCorrector(fileName):
# This method is used for correcting the mistakes or updating the details of the students
    #Split fileName to get name & roll no. of student
    name, rollnumber = fileName.split("_")[:2]
    rollnumber = rollnumber.split(".")[0]
    
    try:   # A 2D list is created which contain Details of the Students
        with open("marksheetData.txt", "r+") as f: 
            data = [line.strip().split(", ") for line in f]
            
    except FileNotFoundError:
        print("Error: Marksheet data file not found!")  # Handling the error if the file is not present
        return
        
    # object is created using Roll No.
    for i in data[2:]:
        if (i[1] == rollnumber):
            student = Student(i)
            break
        
    # Updation menu for the user to choose which of the following user wants to update
    print('''Enter 1 to update Name
Enter  2 to update Programme Name
Enter  3 to Update Marks
Enter 4 to update year of passing\n''')

    try:
        choice1 = int(input("Enter Your Choice: "))
    except ValueError:
        print("Error: Choice Must be Integer!")
        return

    if choice1 == 1:
      newName= input("Enter New Name: ")
      #Student Class object to update name is called
      student.updateName(newName)
      for i in data[2:]:
        if i[1]==rollnumber:
          i[0]=newName  # old name updated
       #os.rename is used to rename the file
      os.rename(f"{name}_{rollnumber}.txt",f"{newName}_{rollnumber}.txt")
      
    elif choice1 ==2:
      newProgramme= input("Enter New Programme Name: ")
      student.updateProgramme(newProgramme)
      for i in data[2:]:
        if i[1]==rollnumber:
          i[2]=newProgramme              #Updating Programme Name
    

    elif choice1 ==3:
      try:
            for i in range(1,6):
                print(f"press {i} to Update Marks of Sub{i}")

            choice2 = int(input("Enter Your Choice: "))

            newMarks = int(input(f" input for New Marks of Sub{choice2}"))

            while(newMarks not in range(101)):
                print("input marks should be Marks should be between 0 & 100")
                newMarks = int(input(f"Please give input for New Marks of Sub{choice2}"))

            student.updateMarks(newMarks,choice2)
            for i in data[2:]:
                if i[1]==rollnumber:
                    i[3+choice2]=str(newMarks)   # Marks update
      except ValueError as e:
            print("Error: ",e)
            return

    elif choice1 ==4:
      try:
          newYear= int(input("Enter New Year: "))
      except ValueError:
          print("Year must be integer ")
          return 
      
          student.updateYear(newYear)
          for i in data[2:]:
            if i[1]==rollnumber:
              i[3]=str(newYear)   # Year update



    with open(student.name+"_"+student.rollno+".txt","w+") as f:
      f.write(student.template())  # update the contents in the Marksheet using write mode
    
    with open('marksheetData.txt', 'w') as f:
    # Write each element(1D List) of the list on a new line
      for line in data:
          f.write(', '.join(line) + '\n')
          #join method accepts a sequence of strings & join all of them with a
          #specified seperator between them (opposite of split())
    print("MarkSheet Updated Successfully!!")
```

```python
def myMSViewer(fileName):
   """
   myMSViewer is a function that show the marksheet of a student.

   I/P(Parameter): fileName - Name of txet file.

   O/P : If file does not exists- 
           FileNotFoundError is raised, & show error messege.
         If file Exists-
           Print the content of file
   Returns: None
   """
   try:                              #handling exception
      with open(fileName,"r") as f: #opening file in reading mode
         print("\n"+f.read()+"\n")       
   except FileNotFoundError:
      print("Error: No Record Found.")  #Print the error messege if file does not exists.
```
